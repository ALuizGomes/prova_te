```python
#fazendos os imports

from PIL import Image
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

  #import dlib
 
import dlib
```

```python
#Lendo as imagens:

imagem1 = cv2.imread('/content/imagem_1.jpg')

#Abrindo a imagem

#cv2_imshow(imagem1)

#Carregando o detector de faces

face_detector = dlib.get_frontal_face_detector()

face_detection = face_detector(imagem1, 1)

#Inserindo os retângulos de detecção

for face in face_detection:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle (imagem1, (l, t), (r, b), (0, 255 , 255), 2)

#Carregando a imagem novamente

cv2_imshow(imagem1)
```

```python
#Carregando o dector de marcos faciais:

face_points_detectors = dlib.shape_predictor('/content/shape_predictor_68_face_landmarks.dat')

#Indentificando os pontos faciais:

for face in face_detection:
  face_points = face_points_detectors = (imagem1, face)

  #Inserindo os pontos:

  for points in face_points.parts():
    cv2.circle(imagem1), (points.x, points.y), 2, (0, 255, 0), 1

  #Inserindo os retângulos novamente:

  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle (imagem1, (l, t), (r, b), (0, 255 , 255), 2)

#Mostrando a Imagem novamente

cv2_imshow(imagem1)
```
